```python
from keras.layers import Conv2DTranspose
from keras.layers.normalization import BatchNormalization
from keras.layers import Activation
from keras import optimizers
```

```python
# Create the Generator

class Generator():

    def __init__(self):
        self.main = Sequential()
        self.main.add(Conv2DTranspose(filters=512, kernel_size=4, strides=(1, 1), padding='valid', use_bias=False))
        self.main.add(BatchNormalization())
        self.main.add(Activation('relu'))
        self.main.add(Conv2DTranspose(filters=256, kernel_size=4, strides=(2, 2), padding='same', use_bias=False))
        self.main.add(BatchNormalization())
        self.main.add(Activation('relu'))
        self.main.add(Conv2DTranspose(filters=128, kernel_size=4, strides=(2, 2), padding='same', use_bias=False))
        self.main.add(BatchNormalization())
        self.main.add(Activation('relu'))
        self.main.add(Conv2DTranspose(filters=64, kernel_size=4, strides=(2, 2), padding='same', use_bias=False))
        self.main.add(BatchNormalization())
        self.main.add(Activation('relu'))
        self.main.add(Conv2DTranspose(filters=3, kernel_size=4, strides=(2, 2), padding='same', use_bias=False))
        self.main.add(Activation('tanh'))
    
    def forward(self, input):
        output = self.main(input)
        return output
        
```
